```python
import cv2
import csv

# Load image
img = cv2.imread('frame0.jpg')

# Define empty list to store ROIs
rois = []


def select_roi(event, x, y, flags, param):
    # Check if left button is pressed down
    if event == cv2.EVENT_LBUTTONDOWN:
        # Record starting point of ROI
        rois.append((x, y))
        
    # Check if left button is released
    if event == cv2.EVENT_LBUTTONUP:
        # Record end point of ROI
        rois.append((x, y))
        # Draw ROI on image
        cv2.rectangle(img, rois[-2], rois[-1], (0, 255, 0), 2)
        # Display image with ROIs
        cv2.imshow('Selector', img)
        
# Create selector window and set mouse callback function
cv2.namedWindow('Selector')
cv2.setMouseCallback('Selector', select_roi)

# Display image with selector window until 'q' is pressed
while True:
    cv2.imshow('Selector', img)
    key = cv2.waitKey(1) & 0xFF
    if key == ord('q'):
        break

# Close all windows
cv2.destroyAllWindows()
```

```python
def drawRectangle(img, a, b, c, d, lowThreshold, highThreshold, min, max):
    sub_img = img[b:b + d, a:a + c]
    if sub_img.size == 0:
        return
    edges = cv2.Canny(sub_img, lowThreshold, highThreshold)
    pix = cv2.countNonZero(edges)

    if pix in range(min, max):
        cv2.rectangle(img, (a, b), (a + c, b + d), (0, 255, 0), 3)
        spots.loc += 1
    else:
        cv2.rectangle(img, (a, b), (a + c, b + d), (0, 0, 255), 3)
```

```python
class spots:
    loc = 0
```

```python
with open('rois.csv', 'r', newline='') as inf:
    csvr = csv.reader(inf)
    rois = list(csvr)

rois = [[int(float(j)) for j in i] for i in rois]
VIDEO_SOURCE = 'D:\smartpark\Khare_testvideo_03.mp4'
cap = cv2.VideoCapture(VIDEO_SOURCE)

while True:
    spots.loc = 0

    ret, frame = cap.read()
    if frame is None:
        break
    
    ret2, frame2 = cap.read()
    min = cv2.getTrackbarPos('Min pixels', 'parameters')
    max = cv2.getTrackbarPos('Max pixels', 'parameters')
    lowThreshold = cv2.getTrackbarPos('Threshold1', 'parameters')
    highThreshold = cv2.getTrackbarPos('Threshold2', 'parameters')

    for i in range(len(rois)-1):
        drawRectangle(frame, rois[i][0], rois[i][1], (rois[i+1][0] - rois[i][0]) , (rois[i+1][1] - rois[i+1][0]), min, max, lowThreshold, highThreshold)
 
    


    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(frame, 'Available spots: ' + str(spots.loc), (10, 30), font, 1, (0, 255, 0), 3)
    cv2.imshow('Detector', frame)

    canny = cv2.Canny(frame2, lowThreshold, highThreshold)
    if canny is not None and canny.shape[0] > 0 and canny.shape[1] > 0:
        cv2.imshow('canny', canny)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    cv2.waitKey(50) 

cap.release()
cv2.destroyAllWindows()
```
